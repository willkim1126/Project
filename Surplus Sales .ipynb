# Surplus Sales Exploratory Data Analysis 

In this notebook, we will look at the sale of Surplus from the dataset called '2.28 Sold Lots.xlsx' which shows all the sales of Surplus store up to 02/28/2024. 

```python
import pandas as pd
import numpy as np
surplus = pd.read_excel('2.28 Sold Lots.xlsx')
surplus
```

## Data Cleaning 

```python
#Basic info about the data 
surplus.shape
```

The code above shows us the dimensionality of the dataframe. We can see that there are 20,379 rows (20,379 entries) and 144 columns (144 attributes for each transaciton). We can easily infer that the dataset is extremely big. Moreover, from first glance of the data, we can see that a lot data are missing we will clean this data to be more concise and usable. 

First, we will drop all the columns that are empty with the code below. 

```python
sp = surplus.dropna(how='all', axis=1)
```

```python
sp.shape
#Shows us that 10 columns were completely empty. 
```

```python
#Sorting the rows by latest transaction date
sp = sp.sort_values(by = 'Transaction Date', ascending = False)
```

We then sort the data by transaction date, starting from the newest transaction. We used 'Transaction Date' as since out of all the dates for each transaction, it seems the most simpliest with most applicable format. 

Still, we have 134 columns of attributes per transaction. It is still obvious that some columns are just mostly NaN or empty. We will drop all the columns that are more than 60% empty. This will allow us to actually run some test and get significant results rather than just null values. 

```python
#Now dropping columns that are more than 60% empty
sp = sp.loc[:, (sp.isin([' ','NULL',0]) | sp.isnull()).mean() <= .6]
```

```python
sp.shape
```

This is what the dataframe looks like so far. It is much more concise and holds meaning data now. 

```python
sp
```

Now, we can see that there are duplicate columns with different column name. Although this might've been important in termns logging all the information about the transaction, in our case, this is just duplicate data that is not useful to us. For example, look at the toy dataframe below. 

```python
toy  = sp[['Lot Category', 'Category']]
toy
```

Some columns hold the same exact data that is not necessary. Now, we will remove duplicate columns of data, leaving only the first occuring column. 

```python
#Now lets remove duplicate columns of data 
sp = sp.T.drop_duplicates().T
```

```python
sp.shape
```

### Columns with String (Text) Data 

For more in-depth look cleaning of the data, we will only look at the columns with the string data right now. 

```python
sp_string = sp.loc[:, sp.iloc[0].map(type) == str]
```
