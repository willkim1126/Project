## Background and Prior Work

Due to the limited availability of on-campus housing at UCSD, commuting has become a routine aspect of the daily lives of many upperclassmen. UCSD students commute from various locations, ranging from across the street and downtown San Diego to Mira Mesa and even from Tijuana. As UCSD students, we were intrigued by whether this seemingly exhausting daily commute has any tangible influence on our academic performance. Furthermore, we sought to determine if the impact of commute time on GPA differs among students with various majors.

There has been a good amount of prior work conducted about this topic, with a lot of interesting results. The department of education conducted a similar study in 2016, where they found that there was a negative correlation between the distance someone has to commute and their GPA. However, they also found that there was a positive relationship between the commute distance squared and the GPA, meaning that students who live exceptionally far away were getting better grades than students who lived a moderate distance away from campus<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). Additionally, another study performed by Gulf Medical University found similar results using a cross sectional study, where there was a negative effect on academic performance due to commuting. They also found that it had a negative effect on a student's physical health, which could be an explanation as to why commuters have worse academic performance<a name="cite_note-2"></a>[<sup>2</sup>](#cite_note-2).

In another study done in the VU University of Amstredam, Netherlands, they analyzed the effect of student commute time on the number of days present, the duration of presence at school, hourly presence per week, and study results.<a name="cite_note-3"></a>[<sup>3</sup>](#cite_note-3) Based on the theoretical framework they’ve created, the test showed that students with longer commute time visit the university often but stay longer, leading to having no effect on weekly hours present in university. However, the study found a negative effect of commute time on overall grade performance. When additional controls, such as province of origin, field of study, and municipality characteristics, are considered, the negative effect becomes more pronounced.

Overall, these studies generally show a negative effect in grade results due to commute distances. In addition to these however, we would also like to see how this effect differs between people from different majors and grade levels, since there has been little to no research done on that specifically. Understanding this could give us a better idea of how commuting can affect certain groups of people differently and make us more aware of the challenges commuters face in an academic setting, especially in UCSD where most of the upperclassmen students have to commute.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import missingno as msno
import statsmodels.api as sm
```

```python
def standardize_study_time(int_in):
    try:
        if int_in == 1:
            output = "0 - 2 hours"
        elif int_in == 2:
            output = "2 - 5 hours"
        elif int_in == 3:
            output = "5 - 10 hours"
        elif int_in == 4:
            output = "More than 10 hours"
        else:
            output = np.nan  
    except:
        output = np.nan

    return output
```

```python
def standardize_commute_time(int_in):
    try:
        if int_in == 1:
            output = "0 - 30 minutes"
        elif int_in ==2:
            output = "30 - 60 minutes"
        elif int_in == 3:
            output = "More than 1 hour"
        else:
            output = np.nan
    except:
        output = np.nan

    return output
```

```python
def standardize_grade_college(float_in):
    if float_in <= 100.00 and float_in >= 81.00:
        output = 'A'
    elif float_in <= 80.99 and float_in >= 61.00:
        output = 'B'
    elif float_in <= 60.99 and float_in >= 41.00:
        output = 'C'
    elif float_in <= 40.99 and float_in >= 0.00:
        output = 'D and F'
    
    return output
```

```python
def standardize_grade_highs(float_in):
    if float_in <= 20.00 and float_in >= 15.50:
        output = 'A'
    elif float_in <= 15.49 and float_in >= 13.50:
        output = 'B'
    elif float_in <= 13.49 and float_in >= 9.50:
        output = 'C'
    elif float_in <= 9.49 and float_in >= 0.00:
        output = 'D and F'
    
    return output 
```

```python
def standardize_gender(str_in):
    if str_in == "Male":
        return "M"
    else:
        return "F"
```

## Dataset #1 : Student Behavior

We are going to use th standardization functions we made above to clean up this dataset to allow us to analyze it.

```python
df_college = pd.read_csv('Student_Behaviour.csv')
```

```python
df_college = df_college[['college mark', 'Travelling Time ', 'daily studing time', 'Gender']]
df_college.columns = ['Grades', 'CommuteTime', 'WeeklyStudyTime', 'Gender']

df_college = df_college.replace(['0 - 30 minutes', '30 - 60 minutes', '1 - 1.30 hour', '1.30 - 2 hour', '2 - 2.30 hour', '2.30 - 3 hour', 'more than 3 hour'], [1,2,3,3,3,3,3])
df_college['CommuteTime'] = df_college['CommuteTime'].apply(standardize_commute_time)
df_college = df_college.replace(['0 - 30 minute', '30 - 60 minute', '1 - 2 Hour', '2 - 3 hour', '3 - 4 hour', 'More Than 4 hour'], [1,2,3,4,4,4])
df_college['WeeklyStudyTime'] = df_college['WeeklyStudyTime'].apply(standardize_study_time)
df_college['Grades'] = df_college['Grades'].apply(standardize_grade_college)
df_college['Gender'] = df_college['Gender'].apply(standardize_gender)
df_college
```

We want to see the grade distributions within each commute time period. To do so, we will create a data frame for each commute time and use matplotlib to plot pie chart.

```python
grade_colors = {'A': 'green', 'B': 'orange', 'C': 'blue', 'D and F': 'red'}
commute_time_order = ["0 - 30 minutes", "30 - 60 minutes", "More than 1 hour"]
```

```python
# Getting unique categories of CommuteTime
unique_commute_times = sorted(df_college['CommuteTime'].unique(), key=lambda x: commute_time_order.index(x))

# Creating subplots for each category of CommuteTime
fig, axes = plt.subplots(1, len(unique_commute_times), figsize=(15, 5))  # Adjust figsize as needed

# Plotting pie charts for each category of CommuteTime
for i, commute_time in enumerate(unique_commute_times):
    # Filter the DataFrame for the current commute time category
    current_commute_time = df_college[df_college['CommuteTime'] == commute_time]
    
    # Calculate percentages for the filtered subset
    grouped_current = current_commute_time['Grades'].value_counts(normalize=True) * 100
    
    # Creating a pie chart within its own subplot
    axes[i].pie(grouped_current, labels=grouped_current.index, autopct='%1.1f%%', startangle=140, colors=[grade_colors.get(grade, 'white') for grade in grouped_current.index])
    axes[i].set_title(f"CommuteTime: {commute_time}")
    axes[i].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.suptitle('Percentage Distribution of Grades for Different CommuteTime Categories', y=1.05)
plt.tight_layout()
plt.show()
```

Based on these pie charts, we can tell that people who have commute times higher than 1 hour tend to have people who get more B's, and D and F's.

We are also going to conduct regression tests to see if there is a relationship between these two variables

```python
# Convert categorical variable CommuteTime to numerical using one-hot encoding
df_encoded = df_college
df_encoded = df_encoded.drop('WeeklyStudyTime', axis=1)
df_encoded = df_encoded.drop('Gender', axis=1)

# Assigning numerical values for Grades using Pandas' categorical codes
grades_mapping = {'A': 0, 'B': 1, 'C': 2, 'D and F' : 3}  # Assign numerical values to grades
df_encoded['Grades'] = df_encoded['Grades'].map(lambda x: grades_mapping[x])
commute_time_mapping = {'0 - 30 minutes': 0, '30 - 60 minutes': 1, 'More than 1 hour': 2}  # Assign numerical values to commute times
df_encoded['CommuteTime'] = df_encoded['CommuteTime'].map(lambda x: commute_time_mapping[x])

# Separating features and target variable
X = df_encoded['CommuteTime']
y = df_encoded['Grades']
```

# Ethics & Privacy

Regarding ethical concerns, our datasets do not mention specifics of where they got their data. Consequently, we are not able to determine whether the sample is random and reflective of our target population. The sample could potentially be a convenience sample, skewing our results and leading to an improper representation of the relationship between grades and commute time. Due to the lack of available datasets, we have to trust the validity of the data. Concerning the privacy-related issues, we found the datasets often provide a plethora of data yet not many specific metrics that could be considered personally identifiable information (PII). The individuals that the data was based off were aware of their data being used as they were from surveys. One of the bias concerns in our datasets is wealth. A student that comes from a wealthy family may be able to afford a higher rent and have a closer residence to campus, reducing their overall commute time. Also, any students that live on campus will have considerably shorter commute times, which we have chosen to address by settings their commute time to 0.

# Discussion and Conclusion

## College students

Our data analysis consisted of three independent variables: commute time, study time, and gender. We then conducted a comparison with corresponding grades. Apart from our main independent variable commute time, we figured it is important to look at other variables as well in order to understand what impacts students' grades as well. Moreover, study time could also be impacted by the duration of commute time, hence worth analyzing. For each comparison, we visualized distribution of grades by groups of independent variables and conducted regression analysis as well. The exploratory data analysis of student behaviors dataset suggests that there is somewhat significant negative correlation between commute time and grades, but not between study time and grades, as well as gender and grades. 

The commute time vs. grades comparison showed that for all three different commute time groups, over 50% of students had B as their grade. Moreover, 21.8% of students with commute time under 30 minutes had A, 26.2% of students with commute time between 30-60 minutes had A, and 15% of students with commute time over 60 minutes had A. The only significant result that we can get out of the pie chart is that once the commute time goes above an hour there is a substantial decrease in academic performance with a decrease in A students. For study time vs. grades and gender vs. grades, there are no notable trends in the distribution, most groups following the same pattern of majority of grades being B and no significant changes among the groups.

After looking at distributions we conducted regression analysis for each comparison. The OLS regression test resulted in 0.524 R-squared for commute time vs. grades, which is statistically significant. Since the R-squared tells us the proportion of variance in the dependent variable that can be explained by independent variables, we can say that around 52.4% of the grades can be explained by their commute time. Other R-squared values from study time vs. grades (0.409) and gender vs. grades (0.112) were both statistically insignificant since it was lower than 0.5. Overall the R-squared values tell us that commute time has statistically significant correlation with grades in the student behaviors dataset. 

Another important measure we got through regression analysis is coefficient term. The commute time coefficient of 0.6667. Important note is that even the OLS regression results gave us 0.6667 which tell us there is a positive relationship between commute time and grades, it is actually the opposite. When we assigned numerical values to grades, we assigned 0 to A, in ascending order. So there is a negative relationship of 0.6667, meaning commute time increases, the grade decreases. For study time vs. grades, the coefficient term was 0.4719, but same as the commute time, the coefficient should be negative due to the numerical conversion of grades. Additionally, for gender vs. grades the coefficient term was 0.7595. We assigned male datasets as 0 and female datasets as 1, showing us that female students had better grades compared to the male students.

## Highschool students

When we began developing our project, our topic of interest was the relationship between academic performance and commute time.  We wanted to see if there was a difference in how commute time impacted academic grades across the education levels of high school and college.  With this in mind we developed a research question, and searched for relevant datasets.  The first dataset we found was regarding student performance for secondary schools in portugal.  The dataset included relevant attributes of traveltime, studytime, gender, and academic grades  for observations of 650 students.   

With this dataset, we were able to conduct our exploratory data analysis, and see if there was a potential correlation between the traveltime, studytime, and grades for the overall students.  When conducting our exploratory data analysis, we wanted to see the distribution of the various attributes, like commute time, study time, grades, and gender, so we created bar plots for each.  We conducted bivariate analysis to see the relationship between different aspects, like Commute Times vs Grade, through a layered barchart. Our plot showed the distribution of letter grades A - F, for each category of commute times, including 0-30 min, 30-60 min and over 1hr. To experiment with visual display, we also plotted used pie charts within each time category. The conclusions we could draw from just visualizations alone were limited, so we conducted regression tests to see if there was a relationship between these two variables. We carried out a similar process to see the relationship between Study Time vs. Grades and Gender vs. Grades.  All visualizations contained a similar distribution of a high percentage of ‘C’ grades, regardless of gender, commute time, and study time. 

From the regression tests, we were able to obtain r squared, which tells how strong the correlation is, and coefficient estimates, which tells what direction the correlation is in, for each of the various predictor variables, like study time, commute time, gender etc.  Based on the r-squared value, we found none of the variables had results over 0.5, which means none yielded significant results to be able to come to a clear conclusion.  The highest r-squared value we obtained was for gender, which was 0.427. Based on the coefficient estimates, we found the highest values were for commute time, which had a coef variable of 1.4322, and gender, which had a coef variable of 1.6475.  Since grades were encoded as 0-5 for A-F respectively, this shows the direction of the relationship, so higher commute time means the grade decreases and since the gender coef is positive. With these results in mind, it was difficult to determine a correlation with any variables. ## Limitations

Our datasets limited the accuracy of our analysis on the grades that the students received. One of our biggest limitations was that the datasets came from two different regions of the world. The Student Behavior Dataset came from a college in India, while the Student Performance Dataset came from a high school in Portugal. Both countries have different grading scales from one another. As a result, we had to fit them into categories ('A', 'B', 'C', 'D and F') according to reported scales from (https://www.scholaro.com/db/Countries/India/Grading-System) for India and (https://en.wikipedia.org/wiki/Academic_grading_in_Portugal) for Portuguese schools. Additionally, we had used an inaccurate grading system initially that inflated the general amount of A's in the Student Behavior dataset. After checking the region of the source, we were able to correct the ranges, giving us a more accurate understanding of the data. Another issue we had was the lack of quantitative data. The data we found in each dataset was mostly categorical, which restricted our analysis models and the visualizations we could make.  This was all a result of a lack of data sets available to us. We were unable to find many datasets that covered our question without making compromises. We were specifically looking for Commute Time and Grades, which not many datasets on Kaggle had. However, we were able to find many similarities between the Student Behavior dataset and the Student Performance dataset. The datasets did not have much information as to how or where specifically they were sourced, which leaves us questioning the validity of our sources.
